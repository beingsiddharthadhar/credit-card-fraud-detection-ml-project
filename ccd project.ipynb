Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

importing the file using pandas

```python
ccd=pd.read_csv("/content/creditcard.csv")
```

Accessing the top 10 data rows of the dataset #amount in

```python
ccd.head(10)
```

```python
ccd.tail()
```

```python
ccd.info()   #dataset info
```

checking the null values in each column

```python
ccd.isnull().sum()
```

```python
ccd['Class'].value_counts()   #0-legit transaction,1-fraudalent transactions
```

separating data for analysis

```python
legit=ccd[ccd.Class==0]
fraud=ccd[ccd.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()      #7947 legit transactions
```

```python
fraud.Amount.describe() #25 fraud transactions
```

compare the values for both transactions

```python
ccd.groupby('Class').mean()
```

Under sampling
Fraud transactions-25
Legit transactions-7947 (dataset is unbalanced)
build a sample dataset of equivalent legit transactions and fraud transactions

```python
legit_sample=legit.sample(n=25)
```

concatenating two dataframe

```python
new_ccd=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_ccd.head()
```

```python
new_ccd.tail()
```

```python
new_ccd['Class'].value_counts()
```

```python
new_ccd.groupby('Class').mean()
```

Splitting the data into features and targets

```python
x=new_ccd.drop(columns='Class',axis=1)
y=new_ccd['Class']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into training data and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model training(we use logistic regression model)

```python
model= LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
#training the logistic regression model with training data
model.fit(x_train,y_train)
```

Model evaluation
accuracy score

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy on training data: ",training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
print("Accuracy on testing data: ",testing_data_accuracy)
```
